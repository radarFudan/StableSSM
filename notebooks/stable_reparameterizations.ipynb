```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# plot function ln(-x) over x = [-5,5], with 1000 points

parameterizations = {}

x = np.arange(-10, 10, 0.01)
parameterizations["vanilla/relu"] = -x
parameterizations["exp"] = -np.exp(x)
parameterizations["softplus"] = -np.log(1+np.exp(x))
parameterizations["sigmoid"] = -1/(1+np.exp(-x))
parameterizations["Best"] = - 1/(x**2 + 0.5)

# Plot ln(-parameterization) for each case
plt.figure(figsize=(6, 5))
for label, y in parameterizations.items():
    # Ensure that we're taking the natural logarithm of negative values
    valid_indices = y < 0
    plt.plot(x[valid_indices], np.log(-y[valid_indices]), label=label)

# Customize the plot
plt.title("")
plt.ylim(-5, 3)
plt.xlabel("w")
plt.ylabel("ln(-f(w))")
plt.legend()
plt.grid(True)

plt.show()
```

```python
parameterizations = {}

x = np.arange(-10, 10, 0.01)
parameterizations["Vanilla/ReLU"] = 1-x
parameterizations["Sigmoid"] = 1/(1+np.exp(-x))
parameterizations["LRU"] = np.exp(-np.exp(x))
parameterizations["Best"] = 1 - 1/(x**2 + 0.5)

plt.figure(figsize=(6, 5))
for label, y in parameterizations.items():
    valid_indices = np.where((y < 1) & (y > -1))
    plt.plot(x[valid_indices], np.log(1-y[valid_indices]), label=label)

# Customize the plot
plt.title("")
plt.ylim(-5, 3)
plt.xlabel(r"$w$")
plt.ylabel("ln(1-f(w))")
plt.legend()
plt.grid(True)

plt.show()
```
